```python
import pandas as pd

poems = pd.read_csv("PoetryFoundationData.csv")
```

```python
def cleanString(row):
    newstr = row
    newstr = ''.join(newstr.split('\r'))
    newstr = newstr.replace('\n\n',' ')
    newstr = newstr.replace('     ',' ;')
    return(newstr.strip())
    
```

```python
toWrite = []
for idx in poems["Poem"]:
    cleaned = cleanString(idx)
    if len(cleaned.split(' ')) > 100 and len(cleaned.split(' ')) < 138:
        toWrite.append(cleaned)
```

```python
print(len(toWrite))
print(toWrite[:30])
```

```python
# Make Corpus:
corpus = open("SmallCorpus.txt","w")
corpus.writelines(toWrite)
```

```python
linesize = []
for l in toWrite:
    linesize.append(len(l.split(' ')))
    
print(linesize)
```

```python
# create encoded texts
#   ~2500 pure
#   1250 single distance + 1250 double distance
#   1250 single distance + 600 double distance
#   1800 word file , 1200 single encoded , 600 double encoded
#   avg size 130 +-10

```

Creating Encoded Texts


```python
import random as rnd
def sliceDump(path):
    fo = open(path)
    encodedDump = fo.read()
    encodedWords = encodedDump.split()
    print(len(encodedWords))
    encodedList = []
    for idx in range(int(len(encodedWords)/130)):
        low = idx*120
        high = (idx+1)*120 + int(rnd.randrange(-15,15))
        #print(low , " " , high)
        if high < len(encodedWords):
            encodedList.append(' '.join(encodedWords[low:high]))
    print(len(encodedList))
    return encodedList
```

```python
unigram = sliceDump("encodedDump.txt")
bigram = sliceDump("encodedDump2.txt")
```

```python
unigram.extend(bigram)

print(len(unigram))

encodedToWrite = unigram[600:3000]
```

```python
for l in encodedToWrite:
    linesize.append(len(l.split(' ')))
print(linesize)
fig,ax = plt.subplots()
line_dist =sb.histplot(linesize,ax=ax,binwidth=2)
ax.set_xlim(100,150)
plt.show()
print(len(toWrite))
```

```python
dataframe = []
#pure
for line in toWrite:
    dataframe.append((line,False))
for line in encodedToWrite:
    dataframe.append((line,True))
```

```python
import pandas as pd
df = pd.DataFrame(dataframe,columns=['text','encoded'])
df = df.sample(frac=1).reset_index(drop=True)

df.to_csv('dataframe.csv')
```
